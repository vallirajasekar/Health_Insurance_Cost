## Data collection through the Various Medium

Policyholder Information: Collect personal details such as age, gender, occupation, and contact information from insurance applicants.

Health Questionnaires: Use health questionnaires to gather data on an individual's medical history, lifestyle choices, habits, and pre-existing conditions.

Medical Records: Access or request medical records from healthcare providers to obtain comprehensive health histories.

Medical Tests: Conduct specific medical tests, including blood tests, cholesterol checks, and BMI assessments, to assess current health status.

Behavioral Data: Utilize wearable devices and health apps to capture data on physical activity, steps taken, and sleep patterns.

Claims Data: Analyze previous insurance claims to understand an individual's medical history and treatment usage.

Third-Party Data: Purchase demographic and regional health data from third-party sources to enhance risk assessment.

Surveys and Market Research: Conduct surveys and market research to gather insights into public health trends and insurance needs.

Telematics and Remote Monitoring: Employ telematics devices and remote health monitoring for real-time behavioral and health data collection.
